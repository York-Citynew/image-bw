```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = plt.imread('china-original.png')
shape = image.shape
```

```python
limage = image.reshape((-1,3))
sorted_centers = np.sort(KMeans(2).fit(limage).cluster_centers_,axis=0)
sorted_centers
```

```python
dists = np.vstack((np.linalg.norm(limage - sorted_centers[0],axis=1),np.linalg.norm(limage - sorted_centers[1],axis=1))).T
labels = np.argmin(dists, axis=1)
labels
```

```python
plt.figure().add_subplot(111, projection='3d')
plt.scatter(limage[:,0],limage[:,1],limage[:,2], c=labels)
```

```python
bw_image = np.vstack((labels,labels,labels)).T.reshape(shape)
plt.imshow(image)
plt.show()
plt.imshow(bw_image * 255)
plt.show()
```
